```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = "../data/raw/MildDemented/mildDem716.jpg"

img = Image.open(image_path)
print("Image size:", img.size)
print("Image mode:", img.mode)
```

```python
plt.imshow(img, cmap="gray")
plt.axis("off")
plt.title("Example Brain MRI")
plt.show()
```

```python
img_array = np.array(img)

print("Array shape:", img_array.shape)
print("Min pixel value:", img_array.min())
print("Max pixel value:", img_array.max())
print("Mean pixel value:", img_array.mean())
```

```python
# Orijinal grnty zaten img deikenine yklemitik
# Gerekirse tekrar:
# from PIL import Image
# img = Image.open("../data/raw/MildDemented/mildDem716.jpg")

import numpy as np

# Grnty NumPy array'e evir
img_array = np.array(img)

print("Original shape:", img_array.shape)
print("Original min:", img_array.min())
print("Original max:", img_array.max())
print("Original mean:", img_array.mean())

# Min-max normalization: [0, 255] -> [0, 1]
img_norm = (img_array - img_array.min()) / (img_array.max() - img_array.min())

print("Normalized min:", img_norm.min())
print("Normalized max:", img_norm.max())
print("Normalized mean:", img_norm.mean())
print("dtype:", img_norm.dtype)
```

```python
import matplotlib.pyplot as plt

plt.imshow(img_norm, cmap="gray")
plt.axis("off")
plt.title("Normalized Brain MRI (0-1 range)")
plt.show()
```

```python
from PIL import ImageOps

# Otomatik kontrast
img_auto = ImageOps.autocontrast(img)

# Histogram equalization
img_eq = ImageOps.equalize(img)
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].imshow(img, cmap="gray")
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(img_auto, cmap="gray")
axes[1].set_title("Auto-contrast")
axes[1].axis("off")

axes[2].imshow(img_eq, cmap="gray")
axes[2].set_title("Equalized")
axes[2].axis("off")

plt.tight_layout()
plt.show()
```
